```python
import urllib.request
from bs4 import BeautifulSoup
import re
```

```python
#обзор на очень плохую игру
html = urllib.request.urlopen('https://kanobu.ru/reviews/retsenziya-na-heavy-rain-373999/')
webpage = html.read().decode('utf-8')
print(webpage)
```

```python
soup = BeautifulSoup (webpage, 'html.parser')
```

```python
#посчитаем сколько раз фамилия создателя игры упоминается всуе
webpage = str(webpage)
print (webpage.count('Кейдж'))
```

```python
#ну и имя
regex = "Дэвид"
allstrings = sorted(re.findall(regex, webpage))
print(len(allstrings))
```

```python
#найдем ссылки на странице
for link in soup.find_all('a'):
    print(link.get('href'))
```

```python
#оставим неповторяющиеся ссылки, ведущие только на внешние ресурсы
l = []
for link in soup.find_all('a'):
    res = link.get('href')
    if res.startswith("http") and res.find("kanobu") == -1:
        if res not in l:
            l.append(res)
for i in l:
  print(i)
```

```python
#найдем  картинки 
for pic in soup.find_all('img'):
    picture = pic['src']
    if picture.find("gif") == -1:
        if picture.find("kanobu") != -1:
            print(picture)
        else: 
            print ("https://kanobu.ru" + picture)
```

```python
#одна картинка потерялась в мета данных
from bs4 import SoupStrainer
only_image = SoupStrainer("meta", property="og:image")
lost_pic = BeautifulSoup(webpage, "html.parser", parse_only = only_image).prettify()
print(re.findall('(?:(?:(?:ftp|http)[s]*:\/\/|www\.)[^\.]+\.[^ \n]+)', lost_pic))
```

```python
#и немного сырого описания автора
script_only = SoupStrainer('script', id = "init_state")
script = BeautifulSoup(webpage, "html.parser", parse_only = script_only).prettify()
a = script.find("author")
d = script.find("social", a)

print(script)
script[a:d]
```

```python
#и встроенные видео нормально

videos = re.findall(r'https?://(?:www\.)?youtube\.com\/(?:watch\?v=[a-zA-Z0-9]*)', webpage)
print(videos)
```
